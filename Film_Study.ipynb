# Feature Engineering

### Categorize ranges of data

```python
lengths = []
for x in df["Runtime_int"]:
    if x < 100:
        lengths.append("short")
    elif 100<=x<165:
        lengths.append("average")
    elif x>=165:
        lengths.append("long")
df["Length_Description"] = lengths
```

```python
df.head(10)
```

```python
# df["Length_Category"] = df.Runtime_int.map(lambda x: 'Short' if x<120 else 'Long')
# df
```

```python
categories = []
for x in df["Released_Year"]:
    if x < 1970:
        categories.append("vintage")
    elif x < 2000:
        categories.append("classic")
    else:
        categories.append("modern")
df["Movie_Era"] = categories
```
